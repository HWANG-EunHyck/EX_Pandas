```python
def set_customFont(font_path):
    from matplotlib import font_manager as fm,rc
 
# 폰트 패밀리 이름 가져오기
    font_name = fm.FontProperties(fname = font_path).get_name()

# 새로운 폰트 패밀리 이름 지정
    rc('font',family = font_name)

set_customFont(r'C:\Windows\Fonts\malgunbd.ttf')
```

```python
plt.plot(DF.loc[('남한',  '합계')],'ro-',markersize =5,label = '남한',linewidth = 2)
plt.plot(DF.loc[('북한',  '합계')],'bo-',markersize =5,label = '북한',linewidth = 2)
plt.xlabel('YEAR')
plt.ylabel('energy')
plt.xticks(rotation =45)
plt.tight_layout()
plt.title('South korea & North korea energy total')
plt.legend()
plt.show()
```

```python
def drawlineGraph(title,columns,fmt,labels,xlabel,ylabel):
    plt.figure(figsize=[14,5])
    for idx,col in enumerate(columns):
        plt.plot(DF.loc[col],fmt[idx],markersize =5,linewidth = 2)

    #for idx,col in enumerate(columns):
    #    plt.plot(DF.loc[col],'ro-',markersize =5,label = f'{labels[idx]}',linewidth = 2)
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation = 45)
    plt.legend(labels)
    plt.show()
```

```python
drawlineGraph('Test',[('남한','수력'),('북한','수력')],['ro-','bo-'],['남한-수력','북한-수력'],'년도','발전량')
```

```python
# - figure 객체 생성
# - 생성된 figure 객체 안에 subplot객체 생성
#   * 그래프 표시 방식 지정 : 가로/세로 몇개 그래프 넣을 지 지정
# - subplot에 그래프를 넣기 
```

```python
# - 하나의 figure에서 수력/화력/원자력/합계 그래프 넣기
fig = plt.figure()

# - add_subplot(행수,열수,번호)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# - 각 subplot에 그래프 넣기
ax1.plot(DF.loc[('남한',  '수력')],'bo-',markersize =5,label = '남한',linewidth = 2)
ax2.plot(DF.loc[('남한',  '화력')],'bo-',markersize =5,label = '남한',linewidth = 2)
ax3.plot(DF.loc[('남한',  '원자력')],'bo-',markersize =5,label = '남한',linewidth = 2)
ax4.plot(DF.loc[('남한',  '합계')],'bo-',markersize =5,label = '남한',linewidth = 2)

# - 각 subplot별로 설정
ax1.set_title('남한 수력발전')
ax1.set_xlabel('연도')
ax1.set_ylabel('수력')

#- 화면에 띄우기
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# - 하나의 figure에서 수력/화력/원자력/합계 그래프 넣기
fig = plt.figure(figsize=(20,5))

# - subplot 객체 생성 : subplots() => axes 객체를 배열에 담아서 반환
#                                 => 행단위로 axes 담아서 반환
axes = fig.subplots(1,4)

energy = ['수력','화력','원자력','합계']
fmts = ['bo-','ro-','ko-','yo-']
for idx,i in enumerate(axes):
        i.plot(DF.loc[('남한',  energy[idx-1])],fmts[idx],markersize =3,label = '남한',linewidth = 1)

# - 전체 창 즉, figure 설정
fig.suptitle('남한 에너지 발전량 변화 동향')

# - 화면에 띄우기
plt.show()
```
