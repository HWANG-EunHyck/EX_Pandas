#### [matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1]
# - 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(line) 그래프
# plot(y축데이터)
plt.plot([1,2,3,4])
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
# plot(x축 데이터, y축 데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께 , 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께 , 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께 , 스타일 등등 지정 
# 색(r,b,y,k,g) 점모양(o,>,<,^,), 선모양('','-','--',':','-.') (linewidth = 2)(markersize = 3)
plt.plot([10,20,30],[5,10,15],'ko-.')
plt.show()
```

```python
# [2] 그래프 그리기 - line(선) 그래프
# plot(x축 데이터, y축 데이터, line2D속성으로 지정)
# line2D의 속성으로 설정 
# - color= 'red',marker = 'o',linestyle = 'dashed',
# - markersize = 10, linewidth =2
# - markeredgcolor => mec = 'blue' 
# - markeredgwidth => mew = 5 
plt.plot([10,20,30],[5,10,15],
         color = 'red',marker = 'o',linestyle = 'dashed',
         markersize = 10, linewidth =4,
         mec = 'blue',mew = 3)
plt.show()
```

```python
# [2] 그래프 두개 그리기 - line(선) 그래프
#plt.plot([10,20,30],[15,25,35], 'bo-')
#plt.plot([10,20,30],[18,29,25], 'ro--')
#plt.plot([10,20,30],[25,25,25], 'k^:')
plt.plot([10,20,30],[15,25,35], 'bo-', [10,20,30],[18,29,25],'ro--', [10,20,30],[25,25,25], 'k^:')
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축,y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성 값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--',label ='V')
plt.plot([10,20,30],[5,30,8],'ro--',label ='2')
plt.title('[MPLIB line2D]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - x축,y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성 값을 범례에서 사용함
#                            loc속성으로 위치 지정 가능 [기:best]
#                            (center) (right) (left)
plt.plot([10,20,30],[15,25,20],'bo--',label ='V')
plt.plot([10,20,30],[5,30,8],'ro--',label ='2')
plt.title('[MPLIB line2D]')
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.legend(loc = 'lower right')
plt.show()
```
