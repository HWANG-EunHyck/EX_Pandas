### DataFrame에서 행/열 추가
- 행 방향 추가
- 열 방향 추가

```python
import pandas as pd
```

```python
# 데이터 준비
data = {'name':['홍길동','마징가','원더우먼'],
        'job':['의적','로봇','영어'],
        'gender':['남','남','여'],
        'age':[100,200,300],
        'height':[180,300,200]}
```

```python
dataDF = pd.DataFrame(data)
```

#### [실습] 행 추가 변수명.loc['추가할행이름'] = 데이터**
    * 변수명.iloc[] ==> X
    * 데이터수 == 컬럼수 경우, 행 추가 가능
    * 데이터수 == 1개 경우, 행 추가 가능
    * 그 외의 경우 , 행 추가 X

```python
##  1개 행 추가 => 데이터 수가 열 개수와 동일한 경우 
dataDF.loc[3] = ['베트맨','박쥐','남',40,190]
```

```python
dataDF.loc[4] = ['베트맨','박쥐','남',40,190]
```

```python
dataDF
```

```python
## 1개 행 추가 => 데이터 수를 1개로
dataDF.loc[5] = None
dataDF
```

```python
## 1개 행 추가 => 데이터 수를 2개로 ==> ERROR
# dataDF.loc[5] = ['스파이더맨','거미']
# dataDF
```

```python
## 1개 행 추가 => 데이터 수를 2개로
dataDF.loc[5] = {'name':'스파이더맨',"job": '거미'}
dataDF
```

#### [실습] 열 추가 변수명['추가할행이름'] = 데이터**
    * 데이터수 == 행 수 경우, 열 추가 가능
    * 데이터수 == 1개 경우, 열 추가 가능
    * 그 외의 경우 , 열 추가 X

```python
# 컬럼/열 추가 => weight : 행 개수 == 데이터 개수
dataDF['weight'] = [90,40,50,60,70,90]
```

```python
dataDF
```

```python
# 컬럼/열 추가 => weight : 1개 == 데이터 개수
dataDF['weight2'] = 0
dataDF
```
